```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('CybersecurityAnalysis.csv')
```

```python
df
```

```python
attack_type_counts = df['Attack Type'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(attack_type_counts, labels=attack_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Number of Cybersecurity Attacks by Type (2015-2024)')
plt.tight_layout()
plt.show()
```

```python
loss_by_attack = df.groupby('Attack Type')['Financial Loss (in Million $)'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.pie(loss_by_attack.values, labels=loss_by_attack.index, autopct='%1.1f%%', startangle=90,
        wedgeprops=dict(width=0.4), pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Average Financial Loss by Attack Type (2015-2024)')
plt.tight_layout()
plt.show()
```

```python
industry_counts = df['Target Industry'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_counts.values, y=industry_counts.index)
plt.title('Number of Attacks by Target Industry (2015-2024)')
plt.xlabel('Number of Attacks')
plt.ylabel('Industry')
plt.tight_layout()
plt.show
```

```python
vulnerability_counts = df['Security Vulnerability Type'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=vulnerability_counts.index, y=vulnerability_counts.values, orient='v', color='green')
plt.title('Number of Attacks by Security Vulnerability Type (2015-2024)')
plt.xlabel('Vulnerability Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
source_counts = df['Attack Source'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=source_counts.index, y=source_counts.values, orient='v', palette='viridis')
plt.title('Number of Attacks by Attack Source (2015-2024)')
plt.xlabel('Attack Source')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
industry_loss = df.groupby('Target Industry')['Financial Loss (in Million $)'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 8))
sns.barplot(x=industry_loss.index, y=industry_loss.values, palette='viridis')
plt.title('Distribution of Financial Loss by Target Industry (2015-2024)')
plt.xlabel('Target Industry')
plt.ylabel('Financial Loss (Million $)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
