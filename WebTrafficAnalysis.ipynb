```python
trafficsourcedf=pd.DataFrame(trafficdf.TrafficSource.value_counts().reset_index())

plt.figure(figsize=(10,6))

ax7=sns.barplot(data=trafficsourcedf, x='TrafficSource', y='count', palette=set2palette)

plt.xlabel('Traffic Source')
plt.ylabel('Count')
plt.title('Count of Sessions from Each Traffic Source')
```

```python
trafficsourcepageviews = trafficdf.groupby('TrafficSource')['PageViews'].mean().round(2).reset_index()

desired_order = ['Organic', 'Paid', 'Referral', 'Social', 'Direct']

trafficsourcepageviews['TrafficSource']=pd.Categorical(
trafficsourcepageviews['TrafficSource'], categories=desired_order, ordered=True
)

plt.figure(figsize=(10,6))

ax8 = sns.barplot(data=trafficsourcepageviews, x='TrafficSource', y='PageViews', palette=set2palette, order=desired_order)

for p in ax8.patches:
    height = p.get_height()
    ax8.annotate(f'{height:.2f}',
                (p.get_x() + p.get_width()/2., height),
                 ha='center', va='bottom',
                 color='black', fontsize=10)

plt.xlabel('Traffic Source')
plt.ylabel('Average Page Views')
plt.title('Average Number of Page Views for Each Traffic Source')
```

```python
trafficsourcesessionduration = trafficdf.groupby('TrafficSource')['SessionDuration'].mean().round(2).reset_index()
trafficsourcesessionduration['TrafficSource']=pd.Categorical(
trafficsourcepageviews['TrafficSource'], categories=desired_order, ordered=True
)

plt.figure(figsize=(10,6))

ax9= sns.barplot(data=trafficsourcesessionduration, x='TrafficSource', y='SessionDuration', palette=set2palette, order=desired_order)

for p in ax9.patches:
    height=p.get_height()
    ax9.annotate(f'{height:.2f}',
                 (p.get_x() + p.get_width()/2., height),
                 ha='center', va='bottom',
                 color='black', fontsize=10)

plt.xlabel('Traffic Source')
plt.ylabel('Average Session Duration')
plt.title('Average Session Duration for Each Traffic Source')
```
